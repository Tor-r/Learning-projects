Соответствующие географические объекты и административно-территориальные единицы на территории современного Санкт-Петербурга отсутствуют, единственным населённым пунктом является сам город Санкт-Петербург.

Создадим список из 30 муниципальных образований Санкт-Петербурга, имеющих статус городов и поселков + сам Петербург.

```python
spb_city = ['Александровская', 'Белоостров', 'Зеленогорск', 'Колпино', 'Комарово', 'Красное Село', 'Кронштадт', 'Левашово', 'Лисий Нос', 'Ломоносов', 'Металлострой', 'Молодёжное', 'Павловск', 'Парголово', 'Песочный', 'Петергоф', 'Петро-Славянка', 'Понтонный', 'Пушкин', 'Репино', 'Санкт-Петербург', 'Сапёрный', 'Серово', 'Сестрорецк', 'Смолячково', 'Солнечное', 'Стрельна', 'Тярлево', 'Усть-Ижора', 'Ушково', 'Шушары']
```

В данных есть пропуски в названии населенных пунктов. Частично убрать пропуски мы можем, посмотрев на колонку center_distance.

В данной колонке, судя по значениям, представлены данные о том, нак каком расстоянии находится объект в Петербурге от центра города (максимально удаленный объект находится в поселке Молодежное - 66 км, и это соответствует действительности). А все пункты, не входящие в город Санкт-Петербург, заполнены NaN. 

Мы проверим эту гипотезу, а пока почти со 100%-й уверенностью можем сказать, что объекты в радиусе 10 км от центра расположены в Санкт-Петербурге. Такие записи отредактируем, остальные (их немного) - удалим из анализа.

```python
# маска для выборки записей
na_spb = (df.locality.isna()) & (df.center_distance < 10000)
print(f'Заменено {len(df[na_spb])} пропущенных значений в названии места')

# присваиваем значение по условию
df.loc[na_spb, 'locality'] = 'Санкт-Петербург'
# удаляем записи
print(f'Удалено {len(df[df.locality.isna()])} записей с пропуском в названии места')
df.dropna(subset=['locality'], inplace=True)
```

Заодно уменьшим количество дубликатов в названиях населенных пунктов, создав колонку с сокращенными именами.

```python
# Функция only_upper() принимает строку и возвращает строку из слов, 
# начинающихся с прописной буквы
def  only_upper(row):
    return re.sub(r'\b[а-я, ё]+\s', "", row['locality']).strip()
```

```python

print(f'Всего уникальных имен населенных пунктов до обработки: {len(df.locality.unique())}')

# с помощью функции only_upper() добавим колонку, в которой будут только чистые имена нас. пунктов
df['locality_short'] = df.apply(only_upper, axis=1)

# df['locality'] = df['locality'].str.replace('поселок городского типа', 'ПГТ')
# df['locality_short'] = df['locality'].str.extract(r"~(\b[а-я, ё]+\s*)") # дорабатывать

print(f'Всего уникальных имен населенных пунктов после обработки: {len(df.locality_short.unique())}')
```

#### **Расстояние до центра**

Проверим как распределены пропуски в столбце center_distance. 

```python
print(f"Количество пропусков в center_distance для локаций, не входящих в СПб: {len(df.query('center_distance.isna() & locality_short not in @spb_city'))}")
print(f"Количество пропусков в center_distance для локаций, входящих в СПб: {len(df.query('center_distance.isna() & locality_short in @spb_city'))}")
```

Как мы и предполагали почти все пропуски в center_distance связаны с тем, что в локациях, не входящих административно в Санкт-Петербург, эта информация не указывается.

В тоже время непосредственно в датасете есть несколько локаций - крупных районов массовой застройки, которые находятся в непосредственной близости к городской черте/КАД, но административно относящихся к Ленобласти. Такие локации в маркетинговых целях позиционируются как квартиры в Санкт-Петербурге. В этой связи имеет смысл обозначить для анализа эти локации как отдельную категорию, также не составит большого труда прописать в отношении них расстояние до центра СПб.

Создадим справочник примерных расстояний до центра (Дворцовая площадь) для наиболее крупных пригородных локаций, используя сервис Яндекс.Карты:

```python
suburb_center_nearest = {'Мурино':15200, 'Кудрово':11600, 'Всеволожск':21000, 'Новое Девяткино':16400, 'Бугры':15800, 'Сертолово':23800, 'Янино-1':14700}
```

```python
# Функция suburbs_dist() принимает датафрейм, находит соответствие значения локации
# в ключе словаря и возвращает значение ключа или возвращает старое значение
def  suburbs_dist(row):
    if row['locality_short'] in suburb_center_nearest:
        return suburb_center_nearest[row['locality_short']]
    else:
        return row['center_distance']
```

```python
df['center_distance'] = df.apply(suburbs_dist, axis=1)
```

```python
df.query('center_distance.isna() & locality_short in @spb_city')
```

Таким образом, ценой небольших трудозатрат мы снизили количество пропусков в center_distance на 35% и обогатили датасет полезной информацией в отношении новых популярных пригородных локаций.

71 пропущенную запись оставим как есть, так как в них есть другая ценная информация.

Добавим в датасет колонку locality_cat с категориями локаций, которые могут пригодится в последующем анализе:

* spb - локации на административной территории Санкт-Петербурга
* suburbs - локации массовой жилой застройки в непосредственной близости к КАД
* oblast - локации в Ленинградской области


```python
# Функция loc_cat() сравнивает содержимое колонки с именами со справочниками и присваивает имя категории
def loc_cat(row):
    if row['locality_short'] in spb_city:
        return 'spb'
    elif row['locality_short'] in suburb_center_nearest:
        return 'suburbs'
    else:
        return 'oblast'
```

```python
df['locality_cat'] = df.apply(loc_cat, axis=1)
df['locality_cat'].value_counts(dropna=False)
```

#### **Факторы стоимости**

Изучим влияние различных факторов на стоимость одного квадратного метра жилья - нашу целевую переменную. 

Для этой цели сделаем выборку наиболее типичных записей. Ранее мы определили, что 6% верхних значений имеют очень большой разброс и в целом сильно выделяются среди остальных записей. Отсечем их из выборки, заодно убрав 1% снизу. Также, изучив получившуюся выборку, считаем целесообразным избавиться от нескольких записей со средней ценой за 1 м2 выше 200 тыс. рублей.

```python
df_cut = df[(df['last_price'] < np.quantile(df.last_price, 0.94)) &
         (df['total_area'] < np.quantile(df.total_area, 0.94)) &
         (df['last_price'] > np.quantile(df.last_price, 0.01)) &
         (df['total_area'] > np.quantile(df.total_area, 0.01))]
df_cut = df_cut[df_cut.m2_price < 200000]
```

```python
# df['total_area'] < np.quantile(df.m2_price, 0.99)
df_cut.m2_price.describe()
```

```python
# средняя цена 1 м2 по выборке
m2_price_av = df_cut['m2_price'].mean()
print(f'Средняя цена 1 м2 по выборке: {m2_price_av: .0f}')
```

```python
# функция вычисляет распределение переменной и отклонение цены м2 от средней
def delta_from_mean(df, col):
    res = df.groupby(by=col)['m2_price'].agg(['mean', 'count'])
    res['delta_from_mean'] = res['mean'] - m2_price_av
    res['%'] = res['delta_from_mean'] / m2_price_av * 100
    return res
```

```python
print('Распределение цены в зависимости от категории этажа')
delta_from_mean(df_cut, 'first_last')
```

```python
print('Распределение цены в зависимости от дня недели размещения объявления')
delta_from_mean(df_cut, 'record_weekd')
```

```python
print('Распределение цены в зависимости от месяца размещения объявления')
delta_from_mean(df_cut, 'record_month')
```

```python
print('Распределение цены в зависимости от года размещения объявления')
delta_from_mean(df_cut, 'record_year')
```

#### **Центр Петербурга**

Изучим как связаны цены на квартиры в Петербурге с близостью к центру города.

```python
# добавим столбец, где округленно укажем расстояние до центра в км
df['km_to_center'] = round(df['center_distance'] / 1000)
```

```python
# создадим выборку объектов из СПб
spb = df.query('locality_short == "Санкт-Петербург"')
```

```python
# выведем средннюю цену для каждого километра
spb_by_km = spb.groupby(by='km_to_center')['m2_price'].mean()
spb_by_km.head(10)
```

```python
spb_by_km.plot(
    
    title='Динамика средней цены м2 по удаленности от центра'
)
plt.show()
```

Из графика мы видим, что с удалением от центра цена квадратного метра квартир резко снижается. Причем имеется некоторый отскок на промежутке от 4 до 7 км.

Мы можем, утверждать в радиусе 7 км (и особенно в радиусе 2 км) от центра города жилье ценится существенно дороже, чем в остальных частях города. Это, по всей видимости, связано с тем, что застройка именно в этих границах воспринимается покупателями как исторический центр Санкт-Петербурга.


```python
# выборка из квартир в центре
spb_center = df.query('locality_short == "Санкт-Петербург" & km_to_center <= 7')
spb_center[["m2_price", "total_area", "rooms", 'ceiling_height', "center_distance"]].describe()
```

```python
spb_center[["m2_price", "total_area", "rooms", 'ceiling_height', "center_distance", 'record_weekd']].hist(figsize = (10, 10), bins = 50)
plt.show()
```

Мы видим, что в центре продаются как очень дорогие, так и очень дешевые квартиры (разброс цен максимальный по выборке). Объяснить это можно тем, что в выборку попали как квартиры в современных элитных комплексах, уникальные видовые квартиры, так и жилье в относительно ветхом фонде, расселенные/расселяемые коммунальные квартиры, прочие объекты, отягощенные грузом проблем. Разница в таких вариантах и дает такой огромный разброс.

В остальных показателях также есть различия с общей выборкой. Потолки в основном не превышают 3-х метров, но и доля высоких квартир значительно больше, чем в среднем. Здесь практически нет студий и открытых планов, а доля однокомнатных квартир существенно ниже чем в среднем по базе. Также видно, что средняя площадь квартир в центре в полтора раза выше, чем по всему региону. А объявления о продаже квартир в центре чаще подают в среду.

```python
# выведем диаграммы рассеивания по выборке центра СПб
sns.pairplot(spb_center[["m2_price", "total_area", "rooms", 'ceiling_height', "center_distance"]])
plt.show()
```

В выборке квартир центра Петербурга фактор расстояния до центра практически не работает, что логично. Более четко выражена связь между увеличением общей площади, количеством комнат, высоты потолков и ценой м2 (отрицательная корреляция).
